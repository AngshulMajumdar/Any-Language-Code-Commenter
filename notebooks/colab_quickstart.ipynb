```python
# BLOCK 2 — Load both LLMs once per runtime
from pathlib import Path
from repo_commenter.llm import TransformersLLMClient
from repo_commenter.commenter import comment_repository, clone_repo, zip_output

generator = TransformersLLMClient("Qwen/Qwen2.5-Coder-1.5B-Instruct")
reviewer = TransformersLLMClient("deepseek-ai/deepseek-coder-1.3b-instruct")
print("Models loaded once.")
```

```python
# BLOCK 3 — Run for a GitHub repo. Re-run only this block for each new repo.
import shutil
from pathlib import Path

GITHUB_URL = input("Paste GitHub repo URL: ").strip()
WORK = Path("/content/code_commenter_work")
INPUT_REPO = WORK / "input_repo"
OUT_REPO = WORK / "commented_repo"
PIPELINE_DIR = WORK / "pipeline_data"
ZIP_PATH = WORK / "commented_repo_full.zip"

if WORK.exists():
    shutil.rmtree(WORK)
WORK.mkdir(parents=True, exist_ok=True)

clone_repo(GITHUB_URL, INPUT_REPO)
report = comment_repository(INPUT_REPO, OUT_REPO, PIPELINE_DIR, generator, reviewer)
zip_output(OUT_REPO, PIPELINE_DIR, ZIP_PATH)
print("Done:", ZIP_PATH)
print("Commented files:", sum(1 for r in report if r["status"] == "commented"))
```
